```python
import os
import time
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()
MODEL = "claude-haiku-4-5-20251001"

print("client ready")

def ask(prompt, temperature):
    response = client.messages.create(
        model=MODEL,
        max_tokens=150,
        temperature=temperature,
        messages=[{"role": "user", "content": prompt}],
    )
    return response.content[0].text

print("ask() defined")
```

```python
prompt = "Write a one-sentence tagline for a coffee shop."

print("=" * 50)
print("TEMPERATURE 0.0  (safe / deterministic)")
print("=" * 50)
for i in range(3):
    print(f"\nRun {i+1}: {ask(prompt, 0.0)}")

print("\n" + "=" * 50)
print("TEMPERATURE 1.0  (creative / random)")
print("=" * 50)
for i in range(3):
    print(f"\nRun {i+1}: {ask(prompt, 1.0)}")
```
